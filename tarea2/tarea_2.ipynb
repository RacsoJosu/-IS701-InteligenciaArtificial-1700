```python
import pandas as pd
```

```python
df = pd.read_csv('unemployment analysis.csv' , sep=",")
```

```python
df_continentes = pd.read_csv('countries of the world.csv', sep=",")
df_paises = pd.read_csv('countries.csv',sep=',')
```

```python
lista_columna_años=df.columns[2:]
```

```python
df_nuevo = pd.DataFrame()
for año in lista_columna_años:
    df_temp2=df.loc[ : , ["Country Name","Country Code",año] ] #estamos iterando año por año y selecionando una a una las columnas d años
    df_temp2['Year']= int(año)
    df_temp2.columns=['Country', 'Country Code', 'Value', 'Year'] # esta renombrando el nombre de las columnas
    df_nuevo = pd.concat([df_nuevo,df_temp2]) # concatenacion de dframe, es como una variable acumuladora para cada uno de los años
```

```python
df_nuevo
```

# Merge

```python
df_unido= pd.merge(left=df_nuevo,right=df_paises,left_on='Country',right_on='Country',how='left')
df_unido
```
