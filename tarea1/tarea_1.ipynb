## Dimesión alumnos

```python
df_alumnos = pd.read_csv('data\data_alumnos.csv', sep=';')
df_alumnos
```

## Dimensión tiempo

```python
df_tiempo= pd.read_csv('data\data_tiempo.csv',sep=';')
df_tiempo
```

## - En archivo JSON modele informacion academica de estudiantes relacionada a hechos como por ejemplo notas 

## tabla de hechos

```python
df_hechos= pd.read_json('data\data_hechos.json')
df_hechos
```

## Mediante el uso de JOINs ( merge ) cree un unico Dataframe

```python
df_resultado = pd.merge(df_alumnos,df_hechos,how="inner")
df_resultado = pd.merge(df_tiempo,df_resultado,how="inner")
df_resultado= df_resultado[['numero_cuenta','nombre','apellido','carrera ','año','periodo','indice_periodo']]
df_resultado
```

## Cree tres diferentes manipulaciones o transformaciones que responda preguntas de negocio (ejemplo: cual es la carrera con mejores promedios a nivel periodo)

## 1.) ¿cuales son los estudiantes con los peores indices de periodo del año 2020 ?
### se desea conocer esta informacion, para poder brindar ayuda si es que es necesaria, para aquellos alumnos y saber la razón de tan bajo desempeño 

```python

df_resultado.sort_values(['indice_periodo'],ascending=True)[df_resultado['año'].isin([2020])][df_resultado['indice_periodo']<65]
```

## 2.) ¿Cuales son los estudiantes que alcanzaron un indice de periodo mayor a 80 en los años 2020 y 2021?
### se desea conocer esta informacion, con el fin de premiar a aquellos estudiantes que han tenido un promedio academico excelente y así poder motivar a los demas estudiantes, para que lo intenten

```python
df_resultado.sort_values(['indice_periodo'],ascending=True)[df_resultado['año'].isin([2020,2021])][df_resultado['indice_periodo']>80]
```

## 3.) ¿cual es la carrera con mejores indices a nivel periodo?

```python
#paso 1: agrupar las filas por carrera
informacion=df_resultado.groupby('carrera ')
```

```python
# paso 2: obtener el dataframe de cada uno de los grupos de las carreras
df_fisica=informacion.get_group('Fisica')
df_matematicas=informacion.get_group('Matematicas')
df_electrica=informacion.get_group('Ing. Electrica')
df_sistemas=informacion.get_group('Ing. En Sistemas')
df_quimica=informacion.get_group('Ing. Quimica')
df_industrial=informacion.get_group('Ing. Industrial')
```

```python
# paso 3: calcular la media aritmetica de cada una las tablas que esta dividida por carrera
# almacenar esos valores en un diccionario
# crear un dataframe con esos valores 
df_medias_indices=pd.DataFrame({'carrera':['Fisica','Matematicas','Ing. Electrica','Ing. En Sistemas','Ing. Quimica','Ing. Industrial'],
                       'media_indices':[df_fisica['indice_periodo'].mean(),df_matematicas['indice_periodo'].mean(),df_electrica['indice_periodo'].mean(),df_sistemas['indice_periodo'].mean(),df_quimica['indice_periodo'].mean(),df_industrial['indice_periodo'].mean()]})
```

```python
# paso 4: mostrar la tabla creada ateriormente
df_medias_indices
```

```python
# paso 5: obtener la carrera con los mejores indices a nivel periodo
# en este caso seria aquella carrera con la media de indices, que sea mayor 
df_medias_indices[df_medias_indices['media_indices']==df_medias_indices['media_indices'].max()]
```
