```python
df_final_1 = pd.DataFrame()
for autor in lista_autores:
    # obtenemos el dataframe de los datos de cada autor
    df_autor = amazon.loc[amazon['Author']==autor,:].copy()
    # la lista de años de trabajo de ese autor
    años = df_autor.loc[:,['Year']].sort_values(['Year'],ascending=True).drop_duplicates().iloc[:,0].values
    #obtenemos el utltimo año que publico y el primero 
    año_max = np.max(años)
    año_min = np.min(años)
    # en este segundo ciclo contabilisamos los registros de cada autor por el nombre de su libro
    # en un determido año para asi, obtener el numero de libros en ese año si no existe sera, 
    # 0 y por lo tanto ese año no trabajo
    for año in range(np.min(años),np.max(años)+1):
        conteo = np.count_nonzero(df_autor.loc[amazon['Year']==año,:'Name'])
        df_temp = pd.DataFrame({
        'Author':[autor],
        'Year':[año],
        'Conteo':[conteo]    
        
        })
        df_final_1=pd.concat([df_final_1,df_temp])
    
```

Que genero por año ha obtenido el mejor rating en promedio. (solo deberia de obtener un registro por año)

- ejemplo de la salida esperada:

```python
pd.DataFrame({
    "Year" : [2009,2010]
    , "Genre" : [ "Fiction" , "Fiction" ]
    , "Mean User Rating": [ 4.59 , 4.61 ]
})
```

```python
# agrupacion de los autores por año para obtener su promedio
df_rating=amazon.groupby(['Genre','Year'],as_index=False).agg({'User Rating':'mean'})
```

```python
# cambio de nombre de las columnas
df_rating.columns = ['Genre', 'Year', 'Mean_User_Rating']
```

```python
# ordenamiento de mayor a menor para asi con las funciones de venatana, 
# marcar con 0 aquellos promedios que sean mayores por año
df_rating=df_rating.sort_values(['Mean_User_Rating'],ascending=False)
```

```python
# uso de las funciones de ventana para diferenciar los registros mayores por año
df_rating['row_number']=df_rating.groupby(['Year']).cumcount()
```

```python
# obtener los registros con uan media mayor
df_final_3=df_rating.loc[df_rating['row_number']==0,:'Mean_User_Rating'].sort_values(['Year'],ascending=True)
```

Podemos ver que son dos años que este usuario ha ido decreciendo

### como necesito los valores de por lo menos dos años consecutivos por autor para asi, determinar
### aquellos que cumplen con la condidicion del ejercicio, hago uso de la funcion shift() para obtener esos valores
### con un argumento negativo para hacer un Lead

```python
# lead de los dos valores siguientes por separado
df_e3['Lead_1']=df_e3.groupby(['Author'],as_index=False)['User Rating'].shift(-1)
df_e3['Lead_2']=df_e3.groupby(['Author'],as_index=False)['User Rating'].shift(-2)
```

```python
# la diferencia de ese lead, con respecto al valor desde el cual dieron "el salto" 
df_e3['diff_1']=df_e3.apply(lambda r: r['Lead_1']-r['User Rating'],axis=1)
df_e3['diff_2']=df_e3.apply(lambda r: r['Lead_2']-r['User Rating'],axis=1)
```

```python
# con la diferencia echa aquellos registros que en la columna diff_1 y diff_2 tengan valores negativos
# son los que tuvieron una baja en el valor de User Rating por lo tanto son los registros o mejor dicho autores que buscamos
df_e3.loc[(df_e3['diff_1']<0)&(df_e3['diff_2']<0),:]
```

```python
# el drop_duplicates es en caso de que alguno, estuviera repetido porque decrecio mas todavia,
# al solo necesitar un autor no necesitamos que este mas de una vez (o bueno asi entiendo el ejercicio)
df_final_3=df_e3.loc[(df_e3['diff_1']<0)&(df_e3['diff_2']<0),['Author']].drop_duplicates()
```

Cuales son los libros que por año se encuentran arriba del percentil 70. (Sin importar genero)

Ejemplo de salida esperada

```python
pd.DataFrame({
    "Year": [2010,2010,2011,2011]
    , "Titulos": ["Libro A","Libro B","Libro Z","Libro Y"]
    , "Autor": ["Autor 1","Autor 2","Autor 1","Autor 3"]
    , "Rating": [4.1, 4.0,4.2,4.3]
    , "Rating .70": [3.8,3.8,3.7,3.7]
})
```

Este ejemplo es para que se haga una idea de la estructura de salida esperada, claramente por año dependiendo de las publicaciones pueden haber varios libros arriba del percentil 70


### bueno por lo que puedo comprender, este ejerecicio se resuelve calculando el percentil 70 por cada año y con ese valor se realizara la comparacion para determinar que valores lo superan 

```python
np.quantile(amazon.loc[:,['User Rating']],0.70)
```

```python
amazon.columns
```

```python
# lista de años
años_= amazon.loc[:,['Year']].sort_values(['Year'],ascending=True).drop_duplicates().iloc[:,0].values
```

```python
años
```

```python
df_final_4 = pd.DataFrame()
for año in años_:
    # obtenemos el dataframe de los autores y libros por año
    df_año = amazon.loc[amazon['Year']==año,:]
    # calculamos el percentil 70
    percentil_70 = np.quantile(df_año.loc[:,['User Rating']],0.70)
   # hacemos la comparacion para obtener los registros que lo superan en ese año
    df_temp= df_año.loc[df_año['User Rating']>percentil_70,['Year','Author','Name','User Rating']].copy()
    # creamos la nueva columna con el valor del percentil
    df_temp['Rating_70'] = percentil_70
    # hacemos la union 
    df_final_4=pd.concat([df_final_4,df_temp])
```
